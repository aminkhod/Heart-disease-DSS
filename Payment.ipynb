```python
df["payment"] = float("NaN")
df["target1"] = float("NaN")
df["target2"] = float("NaN")

for index, row in df_scheme.iterrows():

    if pd.notna(row["hierarchical"]):
        HA = re.findall(r"(>=+)[0-9]+(,<=+)[0-9]+(:)[0-9]+(-->+)[0-9]+", row["hierarchical"])
        HB = re.findall(r"(>=+)[0-9]+(:)[0-9]+(-->+)[0-9]+", row["hierarchical"])
        HC = re.findall(r"(>=+)[0-9]+(,<=)[0-9]+(:)[0-9]+(-->%+)[0-9]+", row["hierarchical"])
        HD = re.findall(r"(>=+)[0-9]+(:)[0-9]+(-->%+)[0-9]+", row["hierarchical"])

        if HA == [('>=', ',<=', ':', '-->')]:
            a = int(re.findall(r">=+([0-9]+),<=+[0-9]+:[0-9]+-->+[0-9]+", row["hierarchical"])[0])
            b = int(re.findall(r">=+[0-9]+,<=+([0-9]+):[0-9]+-->+[0-9]+", row["hierarchical"])[0])
            c = int(re.findall(r">=+[0-9]+,<=+[0-9]+:([0-9]+)-->+[0-9]+", row["hierarchical"])[0])
            d = int(re.findall(r">=+[0-9]+,<=+[0-9]+:[0-9]+-->+([0-9]+)", row["hierarchical"])[0]) * 10

            df.loc[(df["Inc. ID"] == row["Inc. ID"]) & (df["is_qualify"] == 1) & (df["Group"] == "A") &
                   (df["Rides"] >= a) & (df["Rides"] <= b), "payment"] = np.floor((df.loc[:, "Rides"]-a+1)/c) * d
            df.loc[(df["Inc. ID"] == row["Inc. ID"]) & (df["is_qualify"] == 1) & (df["Group"] == "A") &
                   (df["Rides"] > b), "payment"] = np.floor((b-a+1)/c) * d

        if HB == [('>=', ':', '-->')]:
            a = int(re.findall(r">=+([0-9]+):[0-9]+-->+[0-9]+", row["hierarchical"])[0])
            b = int(re.findall(r">=+[0-9]+:([0-9]+)-->+[0-9]+", row["hierarchical"])[0])
            c = int(re.findall(r">=+[0-9]+:[0-9]+-->+([0-9]+)", row["hierarchical"])[0]) * 10

            df.loc[(df["Inc. ID"] == row["Inc. ID"]) & (df["is_qualify"] == 1) & (df["Group"] == "A") &
                   (df["Rides"] >= a), "payment"] = np.floor((df.loc[:, "Rides"]-a+1)/b) * c

        if HC == [('>=', ',<=', ':', '-->%')]:
            a = int(re.findall(r">=+([0-9]+),<=[0-9]+:[0-9]+-->%+[0-9]+", row["hierarchical"])[0])
            b = int(re.findall(r">=+[0-9]+,<=([0-9]+):[0-9]+-->%+[0-9]+", row["hierarchical"])[0])
            c = int(re.findall(r">=+[0-9]+,<=[0-9]+:([0-9]+)-->%+[0-9]+", row["hierarchical"])[0])
            d = int(re.findall(r">=+[0-9]+,<=[0-9]+:[0-9]+-->%+([0-9]+)", row["hierarchical"])[0])

            df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                   (df["is_qualify"] == 1) & (df["Group"] == "A") &
                   (df["Rides"] >= a) &
                   (df["Rides"] <= b), "payment"] = np.floor((df.loc[:, "Rides"]-a+1)/c) * d * df.loc[:, "Commission"]
            df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                   (df["is_qualify"] == 1) & (df["Group"] == "A") &
                   (df["Rides"] > b), "payment"] = np.floor((b-a+1)/c) * d * df.loc[:, "Commission"]

        if HD == [('>=', ':', '-->%')]:
            a = int(re.findall(r">=+([0-9]+):[0-9]+-->%+[0-9]+", row["hierarchical"])[0])
            b = int(re.findall(r">=+[0-9]+:([0-9]+)-->%+[0-9]+", row["hierarchical"])[0])
            c = int(re.findall(r">=+([0-9]+):[0-9]+-->%+([0-9]+)", row["hierarchical"])[0])

            df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                   (df["is_qualify"] == 1) & (df["Group"] == "A") &
                   (df["Rides"] >= a), "payment"] = np.floor((df.loc[:, "Rides"]-a+1)/b) * c * df.loc[:, "Commission"]

    if pd.notna(row["targetbased"]):
        T = re.findall(r">=+(\d+)", row["targetbased"])
        n = len(re.findall(r"(>=+)", row["targetbased"]))
        R = re.findall(r"-->+(\d+)", row["targetbased"])
        T = [int(x) for x in T]
        R = [int(x) * 10 for x in R]

        if n > 1:
            i = 0
            for i in range(0,n):
                if i != n-1:
                    for t in T:
                        df.loc[(df["Inc. ID"] == row["Inc. ID"]) & (df["is_qualify"] == 1) & (df["Group"] == "A") &
                               (df["Rides"] >= T[i]) & (df["Rides"] < T[i+1]), "payment"] = R[i]
                        df.loc[(df["Inc. ID"] == row["Inc. ID"]) & (df["is_qualify"] == 1) & (df["Group"] == "A") &
                               (df["Rides"] >= T[i]) & (df["Rides"] < T[i+1]), "target1"] = 1
                else:
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) & (df["is_qualify"] == 1) &
                           (df["Rides"] >= T[i]) & (df["Group"] == "A"), "payment"] = R[i]
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) & (df["is_qualify"] == 1) &
                           (df["Rides"] >= T[i]) & (df["Group"] == "A"), "target2"] = 1
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) & (df["is_qualify"] == 1) &
                           (df["Rides"] >= T[i]) & (df["Group"] == "A"), "target1"] = float("NaN")
                i += 1
```

```python
for index, row in df_scheme.loc[df_scheme["order of coeffs"].notna(), :].iterrows():
    orders = re.split(r"[/|\\]", row["order of coeffs"])
    for c in orders:
        if c.lower() == "update":
            while True:
                coeff = re.findall(r"^-\++(0.\d+)", row["update coeff."])
                if len(coeff) > 0:
                    coeff = float(coeff[0])
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df["Update"] == 1), "payment"] = df.loc[:, "payment"]*(1+coeff)
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df["Update"] == 0), "payment"] = df.loc[:, "payment"]*(1-coeff)
                    break

                coeff = re.findall(r"^-(0.\d+)", row["update coeff."])
                if len(coeff) > 0:
                    coeff = float(coeff[0])
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df["Update"] == 0), "payment"] = df.loc[:, "payment"] * (1 - coeff)
                    break

                coeff = re.findall(r"^\+(0.\d+)", row["update coeff."])
                if len(coeff) > 0:
                    coeff = float(coeff[0])
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df["Update"] == 1), "payment"] = df.loc[:, "payment"] * (1 + coeff)
                    break

                coeff = re.findall(r"^\+-+(0.\d+)", row["update coeff."])
                if len(coeff) > 0:
                    coeff = float(coeff[0])
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df["Update"] == 1), "payment"] = df.loc[:, "payment"]*(1+coeff)
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df["Update"] == 0), "payment"] = df.loc[:, "payment"]*(1-coeff)
                    break

        if c.lower() == "de": d,r = "Dual","DE coeff."
        if c.lower() == "fake": d,r = "Fake apper","Fake coeff."
        if (c.lower() == "de") | (c.lower() == "fake"):
            while True:
                coeff = re.findall(r"^-\++(0.\d+)", row[r])
                if len(coeff) > 0:
                    coeff = float(coeff[0])
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df[d] == 0), "payment"] = df.loc[:, "payment"]*(1+coeff)
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df[d] == 1), "payment"] = df.loc[:, "payment"]*(1-coeff)
                    break

                coeff = re.findall(r"^-(0.\d+)", row[r])
                if len(coeff) > 0:
                    coeff = float(coeff[0])
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df[d] == 1), "payment"] = df.loc[:, "payment"] * (1 - coeff)
                    break

                coeff = re.findall(r"^\+(0.\d+)", row[r])
                if len(coeff) > 0:
                    coeff = float(coeff[0])
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df[d] == 0), "payment"] = df.loc[:, "payment"] * (1 + coeff)
                    break

                coeff = re.findall(r"^\+-+(0.\d+)", row[r])
                if len(coeff) > 0:
                    coeff = float(coeff[0])
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df[d] == 0), "payment"] = df.loc[:, "payment"]*(1+coeff)
                    df.loc[(df["Inc. ID"] == row["Inc. ID"]) &
                            (df["payment"].notna()) &
                            (df[d] == 1), "payment"] = df.loc[:, "payment"]*(1-coeff)
                    break

Start_Date = df_scheme.loc[0, "From(Date)"]
End_Date = df_scheme.loc[0, "To(Date)"]
Start_Hour = df_scheme.loc[0, "From(hour)"]
End_Hour = df_scheme.loc[0, "To(hour)"]
```

```python
# Create folder for files
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print('Error: Creating directory. ' + directory)


if Start_Date != End_Date:
    directory = './Payment_{} till {}_{} to {}/'.format(Start_Date,
                                                           End_Date,
                                                           Start_Hour[0:2], End_Hour[0:2])
else:
    directory = './Payment_{}_{} to {}/'.format(Start_Date,
                                                   Start_Hour[0:2], End_Hour[0:2])

createFolder(directory)

def excel_exporter(Name, Start_Date, End_Date, Start_Hour, End_Hour):
    if Start_Date != End_Date:
        writer = pd.ExcelWriter(directory + "{}_{} till {}_{} to {}.xlsx".format(Name, Start_Date,
                                                                                        End_Date,
                                                                                        Start_Date,
                                                                                        End_Date,
                                                                                        Start_Hour[0:2], End_Hour[0:2]),\
                                engine='xlsxwriter')
    else:
        writer = pd.ExcelWriter(directory + "{}_{}_{} to {}.xlsx".format(Name, Start_Date,
                                                                                Start_Hour[0:2], End_Hour[0:2],\
                                                                         engine='xlsxwriter'))
    return writer
```
